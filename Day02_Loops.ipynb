# **Chapter 2: Loops in Python**

### Loops allow us to repeatedly execute a block of code until a certain condition is met. There are two types of loops in Python: **for loops** and **while loops.**

![For-vs-While-Loop[1].jpg]()

## **2.1 For Loops**

The for loop iterates over each element in the sequence (here, the list 'fruits') and assigns it to the variable 'fruit'. The code block inside the loop is then executed.

Syntax:

```python
for variable in sequence:
        # code block
```



````python
# Example 01
fruits = ['apple', 'banana', 'cherry',"grapes","mango","kiwi","peach","watermelon"]

for fruit in fruits:
    print("I eat ",fruit)
````

````python
# Example 02

for num in [1, 2, 3, 4, 5]:
    print("Number:", num)
````

````python
# Example

for number in range(0,21,2):
    print("Number:", number)
````

````python
list(range(0,21,2))
````

## **2.2 While Loop**

The while loop repeatedly executes the code block as long as the condition is true. In this example, the loop continues until 'count' is less than 5.

Syntax:

```python
while condition:
    # code block

```

````python
# Example 01

count = 0

while count < 5:
    print("Count:", count)
    count += 1
    # count = count +1
````

````python
#Example 02

password = ""

while password != "Secret123!":
    password = input("Enter the password: ")
    print("Incorrect password. Try again.")

print("Access granted!")
````

````python
# Example 03

num = 10

while num >= 0:
    print("Number:", num)
    num -= 2
````

````python
# Another Example of Infinite Loop

while True:
  print(" I AM STILL RUNNING ")
````

## **2.3 Nested Loops**

Nested loops are loops within loops. They are useful when we need to perform repetitive tasks with multiple levels of iteration.



````python
for i in range(3): #row
    for j in range(4): #col
        print("i =", i, ", j =", j)
````

The outer loop runs three times, and for each iteration, the inner loop runs twice. This creates a matrix-like structure of outputs.

````python
# Example 02

shapes = ['circle', 'square', 'triangle']
colors = ['red', 'green', 'blue']

for shape in shapes:
    for color in colors:
        print(color, shape)
````

````python
# Example 03

for i in range(1, 4):
    for j in range(1, 4):
        print(i * j, end=" ")
    print()
````

## **2.4 Break Statement**

````python
# Example 01

numbers = [1, 2, 3, 5, 6,4,0,9]
for num in numbers:
    if num == 4:
        print("Exiting the loop because num is 4")
        break
    print("Current number:", num)
````

````python
# Example 02

while True:
    password = input("Enter the password: ")
    if password == "secret":
        print("Access granted!")
        break
    else:
        print("Incorrect password. Try again.")
````

````python
numbers = [0,1, 7, 3, 4, 5,6]
for num in numbers:
    if num % 2 == 0:
        break
    print("Current number:", num)
````

### **Conclusion**
